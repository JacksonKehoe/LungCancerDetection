```python
# Height and width are 256, 256
input_tensor = Input(shape=(256, 256, 3))

# Using ResNet50 as the base model; we will make only the last layers trainable
base_model = ResNet50(
    include_top=False, weights=None, input_tensor=input_tensor,
    input_shape=None)
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Binary prediction using sigmoid activation function
x = GlobalAveragePooling2D()(base_model.output)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu', kernel_regularizer=l2(5e-4))(x)
x = Dropout(0.5)(x)

final_output = Dense(1, activation='sigmoid', name='final_output')(x)
model = Model(input_tensor, final_output)
model.summary()
```

```python
# IMG_Set = data flow
STEP_SIZE_VALID = Validation_IMG_Set.n // Validation_IMG_Set.batch_size

opt = SGD(lr=1.0)
model.compile(loss = 'binary_crossentropy', optimizer=opt, 
              metrics=['accuracy'])


#checkpoint_filepath = "models/"

#model_checkpoint_callback = ModelCheckpoint(
#    filepath=checkpoint_filepath,
#    save_weights_only=True,
#    monitor='val_accuracy',
#    mode='max',
#    save_best_only=True)

monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
        patience=10, verbose=1, mode='auto',
        restore_best_weights=True)


history = model.fit(Train_IMG_Set, epochs=32, 
                    validation_data = Validation_IMG_Set, callbacks=[monitor],
                    verbose = 1, validation_steps=STEP_SIZE_VALID)
```

```python
results = model.evaluate(Test_IMG_Set, batch_size=200)
print("test loss, test acc:", results)
```
